```python
import warnings
warnings.filterwarnings("ignore")
import re
import jieba 
import pandas as pd
import numpy as np
import csv
import openpyxl
from sklearn.model_selection import train_test_split
from gensim.models import word2vec
import gensim
```

```python
#去停用词
stopwords = pd.read_csv("./data/stopwords.txt",header=None,quoting = csv.QUOTE_NONE,delimiter="\t")
stopwords = stopwords[0].tolist()
```

```python
def processing(data):
    zw = lambda x: ''.join(re.sub(re.compile(r'[^\u4e00-\u9fa5]'), '', str(x)))
    data['words']=data['内容'].apply(zw)
    data['words']=[jieba.lcut(line) for line in data['words']]
    x1=[]
    for x in data['words']:
        x0=[]
        for word in x:
            if word not in stopwords:
                x0.append(word)
        x1.append(x0)
    for x in range(0,len(data['words'])):
        data['words'].iloc[x]=x1[x]
    return data
```

```python
#加载数据库
pos= pd.read_excel('./data/pos新.xlsx')
neg=pd.read_excel('./data/neg新.xlsx')
pos=processing(pos)
neg=processing(neg)
```

```python
import torch
from torch import nn
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence,pad_packed_sequence
from torch.utils.data import Dataset, DataLoader
import os
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'
device = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
# 超参数
learning_rate = 5e-4
input_size = 300
num_epoches = 10
batch_size = 100
embed_size =300
hidden_size = 64
num_layers = 2
```

```python
#方式一： 数据集
class MyDataset(Dataset):
    def __init__(self, df,df1):
        self.data = []
        self.label = df1.tolist()
        for s in df.tolist():
            vectors = []
            for w in s: 
                if w in model.wv.index2word:
                    vectors.append(model.wv[w])   # 将每个词替换为对应的词向量
            vectors = torch.Tensor(vectors)
            self.data.append(vectors)
    
    def __getitem__(self, index):
        data = self.data[index]
        label = self.label[index]
        return data, label

    def __len__(self):
        return len(self.label)
```

```python
def collate_fn(data):
    """
    param data:第0维：data,第一维：label
    return 序列化的data,记录实际长度的序列，以及label列表
    """
    data.sort(key=lambda x:len(x[0]),reverse=True)# pack_padded_sequence要求要按照序列的长度倒序排列
    data_length = [len(sq[0]) for sq in data]
    x = [i[0] for i in data]
    y = [i[1] for i in data]
    data = pad_sequence(x, batch_first=True, padding_value=0)   # 用RNN处理变长序列的必要操作
    return data, torch.tensor(y, dtype=torch.float32), data_length
```

```python
train_data = MyDataset(xyl_train,yyl_train)
train_loader = DataLoader(train_data, batch_size=batch_size, collate_fn=collate_fn, shuffle=True)
```

```python
test_data = MyDataset(xyl_test,yyl_test)
test_loader = DataLoader(test_data, batch_size=batch_size, collate_fn=collate_fn, shuffle=True)
```

```python
# 网络结构一：
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(LSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size * 2, 1)  # 双向, 输出维度要*2
        self.sigmoid = nn.Sigmoid()

    def forward(self, x, lengths):
        h0 = torch.zeros(self.num_layers * 2, x.size(0), self.hidden_size).to(device)  # 双向, 第一个维度要*2
        c0 = torch.zeros(self.num_layers * 2, x.size(0), self.hidden_size).to(device)
        
        packed_input = torch.nn.utils.rnn.pack_padded_sequence(input=x, lengths=lengths, batch_first=True)
        packed_out, (h_n, h_c) = self.lstm(packed_input, (h0, c0))

        lstm_out = torch.cat([h_n[-2], h_n[-1]], 1)  # 双向, 所以要将最后两维拼接, 得到的就是最后一个time step的输出
        out = self.fc(lstm_out)
        out = self.sigmoid(out)
        return out

lstm = LSTM(embed_size, hidden_size, num_layers)
lstm.to(device)#模型也要放在GPU里，见https://blog.csdn.net/qq_45056135/article/details/125227784
```

```python
from sklearn import metrics

# 在测试集效果检验
def test():
    y_pred, y_true = [], []

    with torch.no_grad():
        for x, labels, lengths in test_loader:
            x = x.to(device)
            outputs = lstm(x, lengths)          # 前向传播
            outputs = outputs.view(-1)          # 将输出展平
            y_pred.append(outputs.cpu()) #转化成cpu，见https://www.cnblogs.com/booturbo/p/16341650.html
            y_true.append(labels.cpu())

    y_prob = torch.cat(y_pred)
    y_true = torch.cat(y_true)
    y_pred = y_prob.clone()
    y_pred[y_pred > 0.5] = 1
    y_pred[y_pred <= 0.5] = 0
    
    print(metrics.classification_report(y_true, y_pred))
    print("准确率:", metrics.accuracy_score(y_true, y_pred))
    print("AUC:", metrics.roc_auc_score(y_true, y_prob) )
    return metrics.accuracy_score(y_true, y_pred)
```

```python
# 定义损失函数和优化器
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(lstm.parameters(), lr=learning_rate)
```

```python
# 迭代训练
acc_=[]
for epoch in range(num_epoches):
    total_loss = 0
    for i, (x, labels, lengths) in enumerate(train_loader):
        x = x.to(device)
        labels = labels.to(device)
        outputs = lstm(x, lengths)          # 前向传播
        logits = outputs.view(-1)           # 将输出展平
        loss = criterion(logits, labels)    # loss计算
        total_loss += loss
        optimizer.zero_grad()               # 梯度清零
        loss.backward(retain_graph=True)    # 反向传播，计算梯度
        optimizer.step()                    # 梯度更新
        if (i+1) % 10 == 0:
            print("epoch:{}, step:{}, loss:{}".format(epoch+1, i+1, total_loss/10))
            total_loss = 0
    
    # test
    acc=test()
    acc_.append(acc)
    
    # save model
    model_path = "./models/lstm_{}.model".format(epoch+1)
    torch.save(lstm, model_path)
    print("saved model: ", model_path)
np.savetxt('./data/plmodelselect.txt',acc_)
```

```python
#网络结构1-2
class BILSTMModel(nn.Module):

    def __init__(self, vocab_size, embedding_dim, pretrained_weight,hidden_dim, n_layers):

        super(BILSTMModel, self).__init__()

        self.hidden_dim = hidden_dim
        self.n_layers = n_layers
        self.embedding = nn.Embedding.from_pretrained(pretrained_weight)       #对于自建单词词向量：nn.Embedding(vocab_size, embedding_dim)词嵌入层：nn.Embedding(num_embeddings, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers, batch_first=False,bidirectional=True)
        self.decoder = nn.Linear(2*hidden_dim,2)
        #self.dropout = nn.Dropout(0.5)#可以加或者不加

    def forward(self, x):
        #这里我的x 形状是batch_size,embedding_dim
        #nn.Embedding(num_embeddings, embedding_dim) embedding的字典大小和每个词用多少维的向量表示
        embedding = self.embedding(x)#必有：提取词特征，输出形状为[batch, seq_len, embed_dim]
        #putput 形状是(seq_len, batch, D*hidden_size)
        output, (final_hidden_state, final_cell_state) = self.rnn(embedding.permute([1, 0, 2]))#必有:rnn.LSTM只返回最后一层的隐藏层在各时间步的隐藏状态。
        #encoding shape: (batch, 2*D*hidden_size)
        encoding = torch.cat([output[0], output[-1]], dim=1)                          
        outs = self.decoder(encoding)                          
        return outs
```

```python
#网络结构二
import torch.nn.functional as F
class BiLSTM_Attention(nn.Module):

    def __init__(self, vocab_size, embedding_dim, pretrained_weight,hidden_dim, n_layers):

        super(BiLSTM_Attention, self).__init__()

        self.hidden_dim = hidden_dim
        self.n_layers = n_layers
        self.embedding = nn.Embedding.from_pretrained(pretrained_weight)       #对于自建单词词向量：nn.Embedding(vocab_size, embedding_dim)词嵌入层：nn.Embedding(num_embeddings, embedding_dim) 
        self.rnn = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers, batch_first=False,bidirectional=True)
        self.decoder = nn.Linear(2*hidden_dim,2)
        #self.dropout = nn.Dropout(0.5)#可以加或者不加

        # 初始时间步和最终时间步的隐藏状态作为全连接层输入
        # What is nn. Parameter(data(Tensor)=None, requires_grad=True) ? Explain
        # 理解为类型转换函数，将一个不可训练的类型Tensor转换成可以训练的类型parameter.
        # 所以经过类型转换这个Tensor变成了模型的一部分，成为了模型中根据训练可以改动的参数了。
        # 使用这个函数的目的也是想让某些变量在学习的过程中不断的修改其值以达到最优化。
        self.w_omega = nn.Parameter(torch.Tensor(hidden_dim * 2, hidden_dim * 2))
        self.u_omega = nn.Parameter(torch.Tensor(hidden_dim * 2, 1))
        #self.decoder2 = nn.Linear(2*hidden_dim, 2)
        nn.init.uniform_(self.w_omega, -0.1, 0.1)
        nn.init.uniform_(self.u_omega, -0.1, 0.1)

    def forward(self, x):
        #这里我的x 形状是batch_size,embedding_dim
        #nn.Embedding(num_embeddings, embedding_dim) embedding的字典大小和每个词用多少维的向量表示
        embedding = self.embedding(x)#必有：提取词特征，输出形状为[batch, seq_len, embed_dim]
        #putput 形状是batch_size, seq_len, 2 * num_hiddens
        output, (final_hidden_state, final_cell_state) = self.rnn(embedding.permute([0, 1, 2]))#必有:rnn.LSTM只返回最后一层的隐藏层在各时间步的隐藏状态。
        #Attention过程
        u = torch.tanh(torch.matmul(output, self.w_omega))         #[batch_size, seq_len, hidden_dim*2]
        att = torch.matmul(u, self.u_omega)
        att_score = F.softmax(att, dim=1)                     #仍为(batch_size, seq_len, 1)
        scored_x = output * att_score                              #[batch_size, seq_len, hidden_dim*2]
        #结束
        context = torch.sum(scored_x, dim=1)                  #[batch_size, hidden_dim*2]
        outs = self.decoder(context)                          ## out形状是(batch_size, 2)
        return outs
```

```python
def test_accuary(test_dataloader):
    """
    @description: get the accuracy of test set
    @param {*}
    - model: Obeject, the Sentiment Analysis model (Seq2Seq).
    - test_dataloader: DataLoader, test DataLoader
    - device: str, choose torch device {"cpu", "cuda"}
    @return {*}
    test_accuracy: the accuracy of total test set
    """
    with torch.no_grad():
        correct = 0
        total = 0
        for k, data_test in enumerate(test_dataloader, 0):
            input_test, target_test = data_test[0], data_test[1]
            input_test = input_test.type(torch.LongTensor)
            target_test = target_test.type(torch.LongTensor)
            target_test = target_test.squeeze(1)
            input_test = input_test.to(device)
            target_test = target_test.to(device)
            output_test = model1(input_test)
            _, pred_test = torch.max(output_test, 1)
            total += target_test.size(0)

            correct += (pred_test == target_test).sum().item()
            F1 = f1_score(target_test.cpu(), pred_test.cpu(), average="weighted")
            Recall = recall_score(target_test.cpu(), pred_test.cpu(), average="micro")
            CM = confusion_matrix(target_test.cpu(), pred_test.cpu())
        print(
            "\nTest accuracy : {:.3f}%, F1_score: {:.3f}%, Recall: {:.3f}%, Confusion_matrix: {}".format(
                100 * correct / total, 100 * F1, 100 * Recall, CM
            )
        )
        return correct / total
```

```python
#如果写出main函数可以参考，https://github.com/SoulDGXu/Sentiment-Analysis-Chinese-pytorch/blob/e94c6ae8f2222e64ab06b9a0cc4997f8b3d5902d/Sentiment_Analysis_main.py
#此处结构为只展示每次迭代训练过程
from sklearn.metrics import confusion_matrix, f1_score, recall_score
import tqdm
def train(train_dataloader):
    acc_=[]
    for epoch in range(num_epochs):
        train_loss,correct,total= 0.0,0.0,0
        train_dataloader = tqdm.tqdm(train_dataloader)
        for i,data_ in enumerate(train_dataloader):
            input_, target = data_[0], data_[1]
            #print(input_.shape)
            #变量移动到GPU
            input_ = input_.type(torch.LongTensor)#必须转化为long，才能nn.embedding
            target = target.type(torch.LongTensor)
            input_ = input_.to(device)
            target = target.to(device)
            #训练
            output = model1(input_)   #模型output
            target = target.squeeze(1)#实际目标label:target, shape:[num_samples, 1]=>[num_samples]
            #print(target.shape)
            #print(output.shape)
            l = loss(output, target)#loss计算
            optimizer.zero_grad()#可以写开头# 梯度清零
            l.backward()# 反向传播，计算梯度
            optimizer.step()# 梯度更新
            train_loss += l.item()#损失记录
            _, predicted = torch.max(output, 1)#按维度dim 返回最大值
            total+=target.size(0)
            correct+=(predicted == target).sum().item() #等价于(predicted.argmax(dim=1) == target).sum().item()
            F1 = f1_score(target.cpu(), predicted.cpu(), average="weighted")
            Recall = recall_score(target.cpu(), predicted.cpu(), average="micro")
            postfix = {"train_loss: {:.5f},train_acc:{:.3f}%"
                ",F1: {:.3f}%,Recall:{:.3f}%".format(
                    train_loss / (i + 1), 100 * correct / total, 100 * F1, 100 * Recall
                )
            }
            
            train_dataloader.set_postfix(log=postfix)#打印日志
        acc = test_accuary(test_dataloader)
        acc_.append(acc)
            # save model
        model_path = "./models/lstm_Attention{}.model".format(epoch+1)
        torch.save(model1, model_path)
        print("saved model: ", model_path)
    np.savetxt('./data/modelselect.txt',acc_)
    return acc_
        
```
